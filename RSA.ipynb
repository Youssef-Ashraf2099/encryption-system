```python
import random
import math

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def generate_prime(bits): # least 256 bits
    while True:
        # Generate a random number with specified bits
        n = random.getrandbits(bits)
        # Check if it's prime
        if is_prime(n):
            return n
        
def SystemModulus(bits):  # least 256 bits
    if bits < 256:
        raise ValueError("Bit length must be at least 256")
    # Generate two distinct prime numbers
    p = generate_prime(bits)
    q = generate_prime(bits)
    
    # Ensure p and q are distinct
    while p == q:
        q = generate_prime(bits)
    
    n = p * q
    
    return n, p, q

def EulerTotient(p, q): #Ã¸(n)
    # Calculate Euler's totient function
    return (p - 1) * (q - 1)

def backSubstitution(a, b, m): # Extended Euclidean Algorithm @MostafaAhmed556 revise keda
    if a == 0 or a==1:
        return (b, 0, 1)
    gcd, x1, y1 = backSubstitution(m % a, a, b)
    x = y1 - (m // a) * x1
    y = x1
    return gcd, x, y

# def modInverse(a, m): # a^-1 mod m
#     gcd, x, y = backSubstitution(a, m, a)
#     if gcd != 1:
#         raise ValueError("Inverse doesn't exist")
#     else:
#         return x % m

#generate public key w kman private make sure men el methods elly fo2 first
```
